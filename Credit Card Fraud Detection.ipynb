# Data Splitting

```python
x = df.drop(columns='Class') #independent Variable
y = df['Class'] #dependent Variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
print('training independent variable data shape',x_train.shape)
print('training dependent variable data shape',y_train.shape)
print('testing independent variable data shape',x_test.shape)
print('test dependent variable data shape',y_test.shape)
```

```python
#feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lg_model = LogisticRegression()
lg_model.fit(x_train,y_train)
```

```python
predict_lg_model =lg_model.predict(x_test) 
predict_lg_model
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
acc = accuracy_score(y_test, predict_lg_model)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, predict_lg_model)
print(cm)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(random_state=0)
classifier_rf.fit(x_train, y_train)
```

```python
y_pred = classifier_rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
acc = accuracy_score(y_test, y_pred)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Gaussian Method

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
Gaussian_predict = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
acc = accuracy_score(y_test, Gaussian_predict)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, Gaussian_predict)
print(cm)
```
